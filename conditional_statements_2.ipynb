````python
"""
write a program to find the even number with below conditions
1.if the number in between 0 to 9 then result is - 1 digit even number
2.if the number in between 10 to 99 then result is - 2 digit even number
3.if the number in between 100 to 999 then result is - 3 digit even number
4.if the number in between 1000 to 9999 then result is - 4 digit even number
5.if other than above even number then the result is - weired even digit number
"""
````

````python
def classify_even_number(num):
if num % 2 != 0:
        return "Not an even number"
if 0 <= num <= 9:
        return "1 digit even number"
elif 10 <= num <= 99:
        return "2 digit even number"
elif 100 <= num <= 999:
        return "3 digit even number"
elif 1000 <= num <= 9999:
        return "4 digit even number"
else:
        return "weired even digit number"
````
